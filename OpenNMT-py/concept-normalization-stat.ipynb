```python
ont='gomf'
#path = './data/concept-normalization-data-craft-v1/'
path = './data/concept-normalization-data-craft-v3-modified-ids/'
```

```python
#training
src_train_data = [i.strip() for i in open(path+ont+'-src-train.txt').readlines()]
tgt_train_data = [i.strip() for i in open(path+ont+'-tgt-train.txt').readlines()]
#testing
src_val_data = [i.strip() for i in open(path+ont+'-src-val.txt').readlines()]
tgt_val_data = [i.strip() for i in open(path+ont+'-tgt-val.txt').readlines()]
```

```python
#frequency of concepts
d = {}
for i in tgt_train_data:
    if i not in d.keys():
        d[i] = 1
    else:
        d[i] += 1
```

```python
sorted_by_value = sorted(d.items(), key=lambda kv: kv[1])
```

```python

for i in reversed(range(len(sorted_by_value))):
    print(sorted_by_value[i][-1])
```
